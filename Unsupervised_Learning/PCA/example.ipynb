```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from PCA import myPCA 
```

```python
from sklearn.datasets import load_wine
```

## Example 1

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
X.shape
```

```python
#plot original data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=iris.target)
plt.title('Original Data')
plt.show()
```

```python
# Initialize your custom PCA class
pca = myPCA()

# Fit the PCA model
pca.fit(X,2)

# Transform the data to reduced dimensions
reduced_data = pca.transform(X)
```

```python
# Reconstruct the original data from the reduced dimensions
reconstructed_data = pca.reconstruct(reduced_data)

# Compute the reconstruction error
reconstruction_error = pca.reconstruction_error_(X)

# Get eigenvalues and eigenvectors
eigenvalues = pca.get_eigenvalues()
eigenvectors = pca.get_eigenvectors()
```

```python
print(f"Variance share: {pca.variance_share:.4f}")
```

```python
print(f"Reduced data shape: {reduced_data.shape}")
print(f"Reconstruction error: {reconstruction_error:.4f}")
```

```python
print(f"eigenvalues: {eigenvalues}")
print(f" eigenvectors:\n{eigenvectors}")
```

```python
# Visualize the first two principal components
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=iris.target)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Iris Dataset - First Two Principal Components")
plt.show()
```

```python
#compare the original data and reconstructed data
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=iris.target)
plt.title('Original Data')
ax = fig.add_subplot(122, projection='3d')
ax.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], reconstructed_data[:, 2], c=iris.target)
plt.title('Reconstructed Data')
plt.show()


```
